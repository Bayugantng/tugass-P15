1.Menghitung laju alir volume minyak

```python
# input
v = float(input("volume = "))
t = float(input("waktu = "))

# proses menghitung aju alir
Q = v/t

# output laju alir flow rate
print(" volume : ", (v), "barel")
print("waktu : ", (t), "jam")
print("v/t = ", (Q), " barel/jam")
```

Codingan ini dibuat untuk menghitung laju alir (flow rate) dalam satuan barel per jam.

Langkah-langkah:
Pengguna diminta memasukkan nilai volume (dalam barel) dan waktu (dalam jam) menggunakan fungsi input. Nilai input diubah menjadi tipe data float menggunakan fungsi float.
Variabel v menyimpan nilai volume dan variabel t menyimpan nilai waktu.

Proses Menghitung Laju Alir:
Laju alir dihitung dengan rumus Q = v / t. Hasil perhitungan disimpan dalam variabel Q.

Output Laju Alir:
Nilai volume, waktu, dan laju alir ditampilkan ke pengguna menggunakan fungsi print.
Format output:
"volume : " diikuti nilai volume (dalam barel)
"waktu : " diikuti nilai waktu (dalam jam)
"v/t = " diikuti nilai laju alir (dalam barel/jam)

Codingan ini sederhana dan mudah dipahami. Pengguna dapat memasukkan nilai volume dan waktu, dan program akan menghitung dan menampilkan laju alirnya.

2.Faktor pemulihan (recovery factor)

```python
# input
Np = float(input("jumlah mniyak yang di produksi = "))
N = float(input("jumlag minyak awal di tempat = "))
m = float(input("gas cap ratio = "))

# proses menghitung faktor pemulihan matrial
RF = Np/N + (m)

# output recovery factor
print("Np : ", (Np), "barel")
print("N : ", (N), "barel")
print("m : ", (m))
print("RF = ", (RF), "meter kubik")
```

Codingan ini dibuat untuk menghitung faktor pemulihan material (RF) minyak di reservoir.

Langkah-langkah:
Pengguna diminta memasukkan nilai:
Jumlah minyak yang diproduksi (Np) dalam barel
Jumlah minyak awal di tempat (N) dalam barel
Rasio tutup gas (gas cap ratio) (m) tanpa satuan
Nilai input diubah menjadi tipe data float menggunakan fungsi float.

Proses Menghitung Faktor Pemulihan Material:
Faktor pemulihan material dihitung dengan rumus RF = Np / N + (m). Hasil perhitungan disimpan dalam variabel RF.

Output Faktor Pemulihan Material:
Nilai Np, N, m, dan RF ditampilkan ke pengguna menggunakan fungsi print.

Format output:
"Np : " diikuti nilai Np (dalam barel)
"N : " diikuti nilai N (dalam barel)
"m : " diikuti nilai m (tanpa satuan)
"RF = " diikuti nilai RF (dalam meter kubik)
Codingan ini menghitung faktor pemulihan material (RF) minyak di reservoir dengan benar. Namun, terdapat kesalahan pada satuan output RF yang seharusnya tidak memiliki satuan.

3.Permeabelitas efektif

```python
# input
krw = float(input("permeabelitas relatif air = "))
sw = float(input("saturasi air = "))
sor = float(input("saturasi minyak residual = "))
n = float(input("eksponen corey = "))
m = float(input("eksponen corey = "))

# proses menghitung pemeabelitas efektif
kr = krw*(sw**n)*(1-sor-sw)**m

# output permeabelitas efektif
print("krw : ", (krw))
print("sw : ", (sw))
print("sor : ", (sor))
print("n : ", (n))
print("m : ", (m))
print("permeabelitas = ", (kr), "meter kubik")
```

Codingan ini dibuat untuk menghitung permeabilitas efektif (kr) batuan reservoir dua fasa air-minyak menggunakan model Corey.

Langkah-langkah:
Pengguna diminta memasukkan nilai:
Permeabilitas relatif air (krw) tanpa satuan
Saturasi air (sw) tanpa satuan
Saturasi minyak residual (sor) tanpa satuan
Eksponen Corey untuk air (n) tanpa satuan
Eksponen Corey untuk minyak (m) tanpa satuan
Nilai input diubah menjadi tipe data float menggunakan fungsi float.

Proses Menghitung Permeabilitas Efektif:
Permeabilitas efektif dihitung dengan rumus kr = krw * (sw**n) * (1 - sor - sw)**m.
Hasil perhitungan disimpan dalam variabel kr.

Output Permeabilitas Efektif:
Nilai krw, sw, sor, n, m, dan kr ditampilkan ke pengguna menggunakan fungsi print.
Format output:
"krw : " diikuti nilai krw (tanpa satuan)
"sw : " diikuti nilai sw (tanpa satuan)
"sor : " diikuti nilai sor (tanpa satuan)
"n : " diikuti nilai n (tanpa satuan)
"m : " diikuti nilai m (tanpa satuan)
"permeabilitas = " diikuti nilai kr (dalam meter kubik)
Codingan ini menghitung permeabilitas efektif (kr) batuan reservoir dua fasa air-minyak menggunakan model Corey dengan benar. Namun, terdapat kesalahan pada satuan output kr yang seharusnya tidak memiliki satuan.

4.Menginvestigasi hubungan harga minyak dengan jumlah produksi minyak

```python
harga_minyak = [55,60,58,62,65,67,70,72,75,77,
                80,82,85,87,90,92,95,97,100,105]
produksi_minyak = [110,120,116,124,130,134,140,144,150,154,
                   160,164,170,174,180,184,190,194,200,210]

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 3))
sns.scatterplot(x=harga_minyak, y=produksi_minyak)
plt.title('Hubungan antara harga minyak mentah per barel')
plt.xlabel('harga_minyak')
plt.ylabel('produksi_minyak')
plt.grid(True)
plt.show()

plt.figure(figsize=(6, 3))
sns.scatterplot(x=harga_minyak, y=produksi_minyak)
plt.title('JUmlah produksi minyak harian di suatu negara')
plt.xlabel('harga_minyak')
plt.ylabel('produksi_minyak')
plt.grid(True)
plt.show()
```

Codingan ini digunakan untuk memvisualisasikan hubungan antara harga minyak mentah per barel dan jumlah produksi minyak harian di suatu negara.

Langkah-langkah:
Import Library:
Import library seaborn dengan alias sns dan matplotlib.pyplot dengan alias plt.

Membuat Data:
Membuat dua list:
harga_minyak: Berisi data harga minyak mentah per barel.
produksi_minyak: Berisi data jumlah produksi minyak harian di suatu negara.

Membuat Visualisasi Scatterplot:
Plot 1:
Gunakan fungsi plt.figure untuk membuat figure dengan ukuran 6 x 4 inci.
Gunakan fungsi sns.scatterplot untuk membuat scatterplot dengan data harga_minyak pada sumbu x dan produksi_minyak pada sumbu y.
Gunakan plt.title untuk memberi judul plot "Hubungan antara harga minyak mentah per barel".
Gunakan plt.xlabel untuk memberi label sumbu x "harga_minyak".
Gunakan plt.ylabel untuk memberi label sumbu y "produksi_minyak".
Gunakan plt.grid(True) untuk menambahkan grid pada plot.
Gunakan plt.show() untuk menampilkan plot.

Plot 2:
Ulangi langkah-langkah yang sama seperti Plot 1, namun ubah judul plot menjadi "Jumlah produksi minyak harian di suatu negara".

Scatterplot digunakan untuk memvisualisasikan hubungan antara dua variabel.
Dalam kasus ini, scatterplot menunjukkan bagaimana harga minyak mentah per barel dan jumlah produksi minyak harian di suatu negara saling terkait.
Semakin tinggi harga minyak, umumnya semakin tinggi pula produksi minyak yang dihasilkan.
Grid pada plot membantu memudahkan pembacaan data.

Codingan ini menghasilkan dua scatterplot yang menunjukkan hubungan antara harga minyak mentah per barel dan jumlah produksi minyak harian di suatu negara. Visualisasi ini dapat membantu memahami pola dan tren dalam data tersebut.

5.Meneiti jumlah rig minyak dan volume ekspor minyak 

```python
jumlah_rig = [15,28,20,22,25,27,30,32,35,37,
              40,42,45,47,50,52,55,57,60,65]
volume_ekspor = [300,360,400,440,500,540,600,640,700,740,
                 800,840,900,940,1000,1040,1100,1140,1200,1300]

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 3))
sns.regplot( x=jumlah_rig, y=volume_ekspor)
plt.title('Hubungan antara jumlah rig minyak yang aktif')
plt.xlabel('jumlah_rig')
plt.ylabel('volume_ekspor')
plt.grid(True)
plt.show()

plt.figure(figsize=(6, 3))
sns.regplot( x=jumlah_rig, y=volume_ekspor)
plt.title('Volume ekspor minyak')
plt.xlabel('jumlah_rig')
plt.ylabel('volume_ekspor')
plt.grid(True)
plt.show()
```

Codingan ini digunakan untuk memvisualisasikan hubungan antara jumlah rig minyak yang aktif dan volume ekspor minyak.
